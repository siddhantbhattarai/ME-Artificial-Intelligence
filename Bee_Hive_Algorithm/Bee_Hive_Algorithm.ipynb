```python
import numpy as np

# Objective function (to minimize)
def objective_function(x):
    return x**2 + 4 * np.sin(5 * x)

# Parameters
num_scout_bees = 10
num_worker_bees = 5
num_onlooker_bees = 5
max_iterations = 100
search_space_bounds = (-10, 10)

# Bee-Hive Algorithm
best_solution = None
best_fitness = float('inf')

for iteration in range(max_iterations):
    scout_positions = np.random.uniform(*search_space_bounds, num_scout_bees)
    
    for scout in scout_positions:
        for _ in range(num_worker_bees):
            local_position = scout + np.random.uniform(-0.5, 0.5)
            local_position = np.clip(local_position, *search_space_bounds)
            fitness = objective_function(local_position)
            if fitness < best_fitness:
                best_fitness = fitness
                best_solution = local_position

    print(f"Iteration {iteration + 1}: Best Solution = {best_solution}, Fitness = {best_fitness}")

print(f"\nFinal Best Solution: {best_solution}, Fitness = {best_fitness}")
```
